# Определение энергии $\gamma$ квантов с помощью сцинтилляционного спектрометра.

## Привожу краткое описание принципа действия спектрометра.

Гамма-кванты от исследуемого источника коллимируются и попадают в сцинтиллятор. Попавшие в кристалл кванты взаимодействуют с его атомами, выбивая из них электроны. Вырванные из атомных оболочек электроны тормозятся в веществе сцинтиллятора, расходуя энергию на возбуждение атомов среды.

Возбужденные атомы высвечиваются, испуская электро-магнитное излучение. Часть образовавшихся фотонов, попадает на катод фотоэлектронного умножителя и в результате фотоэффекта выбивает из него медленные электроны. Эти электроны ускорятся электрическим полем умножителя и вырывают с поверхности первого динода вторичные электроны. Этод щаг многократно повторяется и получается коэффициент умножения порядка $10^6-10^7$. 

При определенном подборе напряжений на динодах величина выходного импульса в широких пределах 
пропорциональна числу фотоэлектронов, вылетевших с фотокатода, а значит и числу фотонов, образовавшихся в сцинтилляторе. Число фотонов, в свою очередь, пропорционально энергии, потерянной в кристалле конверсионными электронами. 

Для определения энергии $\gamma$-квантов источника нужно исследвать кривую распределения их энергетических потерь в кристалле. Идентификацию $\gamma$-линий точнее всего проводить по пикам полного поглощения.

Ширину пика полного поглощения принято характеризовать числом $R$, носящим название энергетического разрешения прибора.

\begin{equation} 
 R = \frac{\sigma}{E}\cdot100\%
\end{equation}

Где величина $\sigma$ определяется как ширина пика полного поглощения на середине высоты ( в ед. энергии); $E$ - энергия регестируемого $\gamma$-излучения.





## Описание полученных результатов:

Измеряю зависимость числа отсчетов в единицу времени от величины порога анализатора при облучении $Co^{60}$.
По графику определяю положение левого фотопика (соответствует энергии в $1.17$ МэВ). 

## Распределение пуассона
Распределение пуассона моделирует случайную величину, представляющую собой число событий, произошедших за фиксированное время, при условии, что данные события происходят с некоторой фиксированной средней интенсивностью и независимо друг от друга.

```python
import matplotlib as mpl
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from matplotlib import mlab
import warnings
warnings.filterwarnings('ignore')
from math import* 
mpl.rcParams['font.size'] = 16

def myFunction(x, a, b, c): 
     return a*x**2+b*x+c
    
def myFunction1(x, a, mu, sigma): 
    return a * e**(-(x - mu)**2 / sigma)   
    
    
data = pd.read_table('inputFile', sep = '\s+' )
x = data['x']
y = data['y']
yErr = [sqrt(elem) for elem in y]
xlist = mlab.frange (x[19], x[28], 0.005)

result = curve_fit(f = myFunction, xdata= x[20:28], ydata=y[20:28], sigma = yErr[20:28],absolute_sigma = True )
popt, pcov = result

result1 = curve_fit(f = myFunction1, xdata= x[20:28], ydata=y[20:28], sigma = yErr[20:28],absolute_sigma = True )
popt1, pcov1 = result1



plt.figure(figsize= (7,7)) 
plt.plot(xlist, myFunction(xlist, *popt), label='Результат фитирования - парабола')
plt.plot(xlist, myFunction1(xlist, *popt1), label='Результат фитирования - Гаусс')
plt.errorbar(x,y,yerr = yErr, fmt='.', label='Экспериментальные точки') 
plt.xlabel("Величина порога анализатора (В).")
plt.ylabel("Число отсчетов в единицу времени.")
plt.grid()
plt.legend();
```
