```python
# Исследование пробега $\alpha$-частиц - счетчик Гейгера.
import matplotlib as mpl
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from matplotlib import mlab
from math import* 

'''
Требуется построить график зависимости N от x, где N - скорость счета, а x - расстояние
между источником и счетчиком.
'''

data = pd.read_table('Geiger', sep = '\s+' )
x = data['x']
y = data['y']
yErr = [sqrt(elem) for elem in y]

plt.figure(figsize= (7,7)) 
plt.errorbar(x,y,yerr = yErr, fmt='.', label='Экспериментальные точки') 
plt.xlabel("расстояние до источника (мм).")
plt.ylabel("Число отсчетов в единицу времени.")
plt.grid()
plt.legend();
```
